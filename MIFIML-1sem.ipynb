3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь

# Функция для преобразования опыта работы в месяцы
def convert_experience_to_months(value):
    # Если значение пропущено или равно "Не указано", возвращаем NaN
    if pd.isna(value) or value == "Не указано":
        return np.nan
    
    # Извлекаем только первую часть строки с годами и месяцами
    experience_part = value.split("  ")[0]
    
    # Инициализируем счетчики лет и месяцев
    years = 0
    months = 0
    
    # Если указаны годы, извлекаем их
    if "год" in experience_part:
        years_part = experience_part.split("год")[0].strip()
        years = int(''.join(filter(str.isdigit, years_part)))  # Извлекаем число
    
    # Если указаны месяцы, извлекаем их
    if "месяц" in experience_part:
        months_part = experience_part.split("месяц")[0].split()[-1].strip()
        months = int(''.join(filter(str.isdigit, months_part)))  # Извлекаем число
    
    # Возвращаем общее количество месяцев
    return years * 12 + months

# Применяем функцию к столбцу "Опыт работы" и создаем новый столбец
df["Опыт работы (месяц)"] = df["Опыт работы"].apply(convert_experience_to_months)
df.drop(columns=["Опыт работы"], inplace=True)

df.head()

```

```python

# Вычисление медианного опыта работы
median_experience = df["Опыт работы (месяц)"].median()

# Вывод результата
print(f"Медианный опыт работы (в месяцах): {median_experience}")
```

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь


# Список городов-миллионников
million_cities = [
    'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск',
    'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 
    'Воронеж', 'Волгоград'
]

# Функция для извлечения категории города
def extract_city(value):
    city = value.split(",")[0].strip()  # Берем первую часть строки (город)
    if city in ["Москва", "Санкт-Петербург"]:
        return city
    elif city in million_cities:
        return "город-миллионник"
    else:
        return "другие"
    
# Функция для определения готовности к переезду
def extract_relocation(value):
    if "готов к переезду" in value or "готова к переезду" in value or "хочу переехать" in value:
        return True
    else:
        return False

# Функция для определения готовности к командировкам
def extract_travel(value):
    if "готов к командировкам" in value or "готова к командировкам" in value:
        return True
    elif "готов к редким командировкам" in value or "готова к редким командировкам" in value:
        return True
    else:
        return False  # Если информации о командировках нет, считаем, что не готов

# Создаем новые признаки
df["Город"] = df["Город, переезд, командировки"].apply(extract_city)
df["Готовность к переезду"] = df["Город, переезд, командировки"].apply(extract_relocation)
df["Готовность к командировкам"] = df["Город, переезд, командировки"].apply(extract_travel)

# Удаляем исходный столбец
df.drop(columns=["Город, переезд, командировки"], inplace=True)

df.head()
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь

# Получаем уникальные категории для "Занятость" и "График"
employment_types = set()
schedule_types = set()

# Извлекаем уникальные категории из столбца "Занятость"
df["Занятость"].dropna().apply(lambda x: employment_types.update([item.strip() for item in x.split(",")]))
print("Уникальные категории в столбце 'Занятость':", employment_types)

# Извлекаем уникальные категории из столбца "График"
df["График"].dropna().apply(lambda x: schedule_types.update([item.strip() for item in x.split(",")]))
print("Уникальные категории в столбце 'График':", schedule_types)

# Преобразуем категории "Занятость" в One Hot Encoding
for employment in employment_types:
    df[employment] = df["Занятость"].apply(lambda x: employment in x if pd.notna(x) else False)

# Преобразуем категории "График" в One Hot Encoding
for schedule in schedule_types:
    df[schedule] = df["График"].apply(lambda x: schedule in x if pd.notna(x) else False)

# Удаляем исходные столбцы
df.drop(columns=["Занятость", "График"], inplace=True)

df.head()
```

```python
# Подсчет количества соискателей, проживающих в Санкт-Петербурге
spb_count = df[df["Город"] == "Санкт-Петербург"].shape[0]

# Общее количество соискателей
total_count = df.shape[0]

# Вычисление процента соискателей из Санкт-Петербурга
spb_percentage = (spb_count / total_count) * 100

# Округление до целого числа
spb_percentage = round(spb_percentage)

# Вывод результата
print(f"Процент соискателей, проживающих в Санкт-Петербурге: {spb_percentage}%")

# Подсчет количества соискателей, готовых к переездам и командировкам
relocation_and_travel_count = df[(df["Готовность к переезду"] == True) & (df["Готовность к командировкам"] == True)].shape[0]

# Общее количество соискателей
total_count = df.shape[0]

# Вычисление процента соискателей, готовых к переездам и командировкам
relocation_and_travel_percentage = (relocation_and_travel_count / total_count) * 100

# Округление до целого числа
relocation_and_travel_percentage = round(relocation_and_travel_percentage)

# Вывод результата
print(f"Процент соискателей, готовых одновременно к переездам и командировкам: {relocation_and_travel_percentage}%")

# Подсчет количества людей, которые ищут проектную работу и волонтерство
project_and_volunteer_count = df[(df["проектная работа"] == True) & (df["волонтерство"] == True)].shape[0]

# Вывод результата
print(f"Количество людей, которые ищут проектную работу и волонтерство: {project_and_volunteer_count}")

# Подсчет количества людей, которые хотят работать вахтовым методом и с гибким графиком
shift_and_flexible_count = df[(df["вахтовый метод"] == True) & (df["гибкий график"] == True)].shape[0]

# Вывод результата
print(f"Количество людей, которые хотят работать вахтовым методом и с гибким графиком: {shift_and_flexible_count}")
```

В чем наша беда? В том, что помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать, например:
* 30000 руб.
* 50000 грн.
* 550 USD

Нам бы хотелось видеть заработную плату в единой валюте, например, в рублях. Возникает вопрос, а где взять курс валют по отношению к рублю?

На самом деле язык Python имеет в арсенале огромное количество возможностей получения данной информации, от обращения к API Центробанка, до использования специальных библиотек, например pycbrf. Однако, это не тема нашего проекта.

Поэтому мы пойдем в лоб: обратимся к специальным интернет-ресурсам для получения данных о курсе в виде текстовых файлов. Например, MDF.RU, данный ресурс позволяет удобно экспортировать данные о курсах различных валют и акций за указанные периоды в виде csv файлов. Мы уже сделали выгрузку курсов валют, которые встречаются в наших данных за период с 29.12.2017 по 05.12.2019. Скачать ее вы можете **на платформе**

Создайте новый DataFrame из полученного файла. В полученной таблице нас будут интересовать столбцы:
* "currency" - наименование валюты в ISO кодировке,
* "date" - дата, 
* "proportion" - пропорция, 
* "close" - цена закрытия (последний зафиксированный курс валюты на указанный день).


Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Пропорция - это число, за сколько единиц валюты указан курс в таблице с курсами. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь

# Преобразуем столбец "Обновление резюме" в формат даты (DD.MM.YYYY HH:MM)
df["Обновление резюме"] = pd.to_datetime(df["Обновление резюме"], format='%d.%m.%Y %H:%M')
# Преобразуем дату из столбца "Обновление резюме" в формат без времени (только дата)
df["Обновление резюме"] = df["Обновление резюме"].dt.date

# Чтение файла с курсами валют
exchange_rates = pd.read_csv('ExchangeRates.csv')
# Преобразуем столбец "date" в формат даты (DD/MM/YY)
exchange_rates["date"] = pd.to_datetime(exchange_rates["date"], format='%d/%m/%y')
# Преобразуем дату в таблице курсов валют в формат без времени (только дата)
exchange_rates["date"] = exchange_rates["date"].dt.date


# Функция для извлечения валюты из столбца "ЗП"
def extract_currency(value):
    if pd.isna(value):
        return None
    currency_map = {
        "грн": "UAN",
        "грн.": "UAN",
        "USD": "USD",
        "белруб": "BYN",
        "бел.руб": "BYN",
        "бел.руб.": "BYN",
        "EUR": "EUR",
        "KZT": "KZT",
        "KGS": "KGS",
        "сум": "UZS",
        "руб.": "RUB",
        "AZN": "AZN"   
    }
    currency = value.split()[-1].strip()
    v=currency_map.get(currency, None)
    if v == None:
        print(value)
    return currency_map.get(currency, None)

# Функция для извлечения суммы из столбца "ЗП"
def extract_salary(value):
    if pd.isna(value):
        return None
    return int(value.split()[0].replace(" ", "").strip())


# Создаем новые столбцы для валюты и суммы
df["currency"] = df["ЗП"].apply(extract_currency)
df["salary"] = df["ЗП"].apply(extract_salary)

# Объединяем данные с курсами валют
df = pd.merge(
    df,
    exchange_rates,
    how="left",
    left_on=["currency", "Обновление резюме"],
    right_on=["currency", "date"]
)

# Рассчитываем зарплату в рублях
def calculate_salary_rub(row):
    if row["currency"] == "RUB":
        return row["salary"]  # Если валюта рубли, возвращаем зарплату без изменений
    elif pd.notna(row["close"]) and pd.notna(row["proportion"]):
        return row["salary"] * row["close"] / row["proportion"]
    else:
        return None  # Если нет данных о курсе валюты, оставляем NaN

df["ЗП (руб)"] = df.apply(calculate_salary_rub, axis=1)
# Удаляем исходные и промежуточные столбцы
df.drop(columns=["ЗП", "currency", "salary", "date", "proportion", "close", "time", "vol", "per"], inplace=True)

df.head()

```

```python
# Вычисление медианной заработной платы
median_salary = df["ЗП (руб)"].median()

# Вывод результата
print(f"Медианная желаемая заработная плата (в рублях): {median_salary}")

# Вычисление модального значения возраста
mode_age = df["Возраст"].mode()[0]

# Вывод результата
print(f"Модальное значение возраста соискателей: {mode_age}")

# Вычисление максимального опыта работы
max_experience = df["Опыт работы (месяц)"].max()

# Вывод результата
print(f"Максимальный опыт работы (в месяцах): {max_experience}")

# Группировка данных и расчет медианной заработной платы по категориям образования
median_salary_by_education = df.groupby("Образование")["ЗП (руб)"].median()

# Поиск категории с максимальной медианной заработной платой
highest_paid_education = median_salary_by_education.idxmax()
highest_median_salary = median_salary_by_education.max()

# Вывод результата
print(f"Категория образования с самой высокой медианной зарплатой: {highest_paid_education}")
print(f"Медианная зарплата для этой категории: {highest_median_salary} руб.")

# Поиск строки с максимальной заработной платой
max_salary_row = df[df["ЗП (руб)"] == df["ЗП (руб)"].max()]

# Извлечение города с максимальной заработной платой
city_with_max_salary = max_salary_row["Город"].values[0]
max_salary = max_salary_row["ЗП (руб)"].values[0]

# Вывод результата
print(f"Город с наибольшим показателем желаемой заработной платы: {city_with_max_salary}")
print(f"Наибольшая желаемая заработная плата: {max_salary} руб.")

# Фильтрация соискателей, готовых и к переезду, и к командировкам
relocation_and_travel_df = df[(df["Готовность к переезду"] == True) & (df["Готовность к командировкам"] == True)]

# Вычисление медианной заработной платы
median_salary = relocation_and_travel_df["ЗП (руб)"].median()

# Вывод результата
print(f"Желаемая медианная заработная плата соискателей, готовых и к переезду, и к командировкам: {median_salary} руб.")

# Группировка данных и расчет средней заработной платы и среднего опыта работы
education_stats = df.groupby("Образование").agg(
    avg_salary=("ЗП (руб)", "mean"),
    avg_experience=("Опыт работы (месяц)", "mean")
)

# Рассчитываем интенсивность роста заработной платы
education_stats["growth_rate"] = education_stats["avg_salary"] / education_stats["avg_experience"]

# Находим категорию с наибольшей интенсивностью роста
fastest_growth_education = education_stats["growth_rate"].idxmax()
fastest_growth_rate = education_stats["growth_rate"].max()

# Вывод результата
print("Категория образования с самым быстрым карьерным ростом:", fastest_growth_education)
print(f"Интенсивность роста заработной платы: {fastest_growth_rate:.2f} руб./месяц опыта")
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# ваш код здесь
# Удаляем строки с пропущенными значениями в "Возраст"
df = df.dropna(subset=["Возраст"])

# Преобразуем возраст в логарифмический масштаб
df["log_Возраст"] = np.log(df["Возраст"])

# Рассчитываем статистики для метода трех сигм
log_age_mean = df["log_Возраст"].mean()
log_age_std = df["log_Возраст"].std()
lower_bound = log_age_mean - 3 * log_age_std
upper_bound = log_age_mean + 4 * log_age_std  # Послабление на 1 сигму вправо

# Построение распределения в логарифмическом масштабе
plt.figure(figsize=(10, 6))
histplot = sns.histplot(df["log_Возраст"], kde=True, color="skyblue", bins=30)
histplot.axvline(log_age_mean, color='k', lw=2, label="Среднее")
histplot.axvline(lower_bound, color='r', linestyle="--", lw=2, label="-3 сигмы")
histplot.axvline(upper_bound, color='g', linestyle="--", lw=2, label="+4 сигмы")
plt.title("Распределение возраста в логарифмическом масштабе")
plt.xlabel("Логарифм возраста")
plt.ylabel("Частота")
plt.legend()
plt.grid(alpha=0.5)
plt.tight_layout()
plt.show()

# Поиск выбросов с помощью метода z-отклонения
z_scores = (df["log_Возраст"] - log_age_mean) / log_age_std
outliers = df[(z_scores < -3) | (z_scores > 4)]

# Удаление выбросов из данных
df_cleaned = df[~((z_scores < -3) | (z_scores > 4))]

# Вывод выбросов
print("Выбросы по возрасту:")
print(outliers[["Возраст", "log_Возраст"]])

# Оценка возраста выбросов
print("\nВозраст соискателей, попавших в категорию выбросов:")
print(outliers["Возраст"].unique())
```

Соискатели, имеющие возраст 14,15 и 100 лет являются выбросами и следует исключить их из рассматриваемых данных как аномалии.
